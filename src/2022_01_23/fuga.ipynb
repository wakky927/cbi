```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

from modules import image
```

```python
im1 = cv2.imread("../../data/2022_01_23/calib/_00002000.bmp", 0)
im2 = np.zeros_like(im1)
```

```python
fig = plt.figure(figsize=(32, 20))
plt.imshow(im1, cmap="gray")
plt.show()
```

```python
src_pts = np.array([[291, 301], [303, 902], [1024, 855], [1027, 306]], dtype=np.float32)
dst_pts = np.array([[290, 376], [299, 881], [827, 842], [830, 381]], dtype=np.float32)
mat = cv2.getPerspectiveTransform(src_pts, dst_pts)
```

```python
mat
```

```python
perspective_img = cv2.warpPerspective(im1, mat, (1280, 1024))
```

```python
fig = plt.figure(figsize=(32, 20))
plt.imshow(perspective_img, cmap="gray")
plt.show()
```

```python
cv2.imwrite('perspective_dst.jpg', perspective_img)
```
