```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("q/rbi_q_1_0505.CSV", header=0, skiprows=56, skipfooter=3, usecols=range(1, 7), engine='python')
df.columns = ["日時 [μs]", "P1 [kPa]", "P2 [kPa]", "Valve [V]", "Flow rate [L/min]", "Air Temp [℃]"]
```

```python
df["Flow rate [L/min]"][202:502].mean()
```

```python
x = [i * 10 for i in range(0, 300)]
m = [df["Flow rate [L/min]"][202:502].mean() for _ in range(0, 300)]

fig, ax = plt.subplots(figsize=(50, 10))
plt.xlabel("$t$ [ms]", fontsize=100)
plt.ylabel("$Q$ [L/min]", fontsize=100)
plt.xticks(fontsize=100)
plt.yticks(fontsize=100)
plt.xlim(-10, 3005)
plt.ylim(0, 5.5)
plt.plot(x, df["Flow rate [L/min]"][202:502], lw=10, c="orange")
plt.plot(x, m, lw=7.5, c="orange", linestyle='dashed')
plt.grid(True)
plt.savefig("1_0505.png", dpi=80)
plt.show()
```
