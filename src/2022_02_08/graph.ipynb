```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
ave_dx = np.loadtxt("../../data/1/dx_0_1.csv", delimiter=',')
ave_dy = np.loadtxt("../../data/1/dy_0_1.csv", delimiter=',')
```

```python
width, height = 739, 692
n = 48

x, y = np.meshgrid(np.linspace(0, width, n, dtype="int"), np.linspace(0, height, n, dtype="int"))
```

```python
fig, ax = plt.subplots(figsize=(12, 10))
c = np.sqrt(ave_dx**2 + ave_dy**2)
im = ax.quiver(x, y, -ave_dx, ave_dy, c, cmap="jet")
ax.invert_xaxis()
fig.colorbar(im)
im.set_clim(0, 1)
plt.show()
```
