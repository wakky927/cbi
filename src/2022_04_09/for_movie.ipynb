```python
SUPER = "/Users/loblium/tmp/cbi/"
IMAGE_DIR = "C001H001S0001_/"
CSV_DIR = "C001H001S0001/"
OUTPUT_DIR = "output/"
```

```python
for i in range(1001, 2001):
    img = cv2.imread(SUPER + IMAGE_DIR + f"C001H001S0001{i:06}.bmp")
    cv2.rectangle(img, (90, 290), (820, 910), (255, 144, 30), 5)
    cv2.imwrite(SUPER + OUTPUT_DIR + f"C001H001S0001{i:06}.bmp", img)
```

### img 2001 to 5000 -> 検出したところを赤で囲う

```python
for i in range(2001, 5000):
    img = cv2.imread(SUPER + IMAGE_DIR + f"C001H001S0001{i:06}.bmp")
    cv2.rectangle(img, (90, 290), (820, 910), (255, 144, 30), 5)
    data = np.loadtxt(SUPER + CSV_DIR + f"{i:06}.csv", delimiter=',', dtype=int, skiprows=1)
    if data.size == 0:
        continue
    elif data.ndim == 1:
        cv2.rectangle(img, (data[0]+90, data[1]+290), (data[0]+90+20, data[1]+290+20), (0, 0, 255), 2)
    else:
        for a, b in data:
            cv2.rectangle(img, (a+90, b+290), (a+90+20, b+290+20), (0, 0, 255), 2)
    cv2.imwrite(SUPER + OUTPUT_DIR + f"C001H001S0001{i:06}.bmp", img)
```

```python
plt.figure(figsize=(16, 16))
plt.imshow(img)
plt.show()
```
